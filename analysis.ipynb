### Automated Exploratory Data Analysis

```python
import pandas as pd

# Load dataset
df = pd.read_csv("steam_reviews 1.csv")

# Clean missing values if any
df = df.dropna(subset=['review'])

# Aggregate data to find hidden commercial value
game_evaluation = df.groupby('title').agg(
    total_reviews=('review', 'count'),
    median_hours=('hour_played', 'median'),
    rec_rate=('recommendation', lambda x: (x == 'Recommended').sum() / len(x) * 100),
    loyal_player_pct=('hour_played', lambda x: (x > 100).mean() * 100)
).reset_index()

# Define an "Acquisition Priority Score" (Weighted Index)
# 50% Sentiment Quality, 50% Player Retention (% over 100 hours played)
game_evaluation['acquisition_score'] = (game_evaluation['rec_rate'] * 0.5) + (game_evaluation['loyal_player_pct'] * 0.5)

# Sort by highest commercial opportunity
top_acquisitions = game_evaluation.sort_values(by='acquisition_score', ascending=False)
print(top_acquisitions[['title', 'median_hours', 'rec_rate', 'loyal_player_pct', 'acquisition_score']].head(10))
```
